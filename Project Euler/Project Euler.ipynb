### Sum square difference
The sum of the squares of the first ten natural numbers is,

1^2 + 2^2 + ... + 10^2 = 385

The square of the sum of the first ten natural numbers is,

(1 + 2 + ... + 10)^2 = 55^2 = 3025

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 âˆ’ 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
n = 100
number_range = np.arange(1, n+1)

def sumsq(n):
    sumsq = 0
    for i in number_range:
        sumsq+=i**2
    return sumsq
def sqsum(n):
    sumrange = 0
    for i in number_range:
        sumrange+=i
    sqsum = sumrange**2
    return sqsum

difference = sqsum(n)-sumsq(n)

print(f"Sum of Squares: {sumsq(n)}\nSquare of Sums: {sqsum(n)}\nDifference: {difference}")
```

## Problem 7
### 10001st prime

By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10 001st prime number?

```python
# This is called the Sieve of Earatosthenes
# See https://stackoverflow.com/questions/567222/simple-prime-generator-in-python
# for some documentation
# There is another way to do it if you look at Project Euler solution

limit = 1000000
desired = 10001
primes=[]
def gen_primes():
    D = {}
    q = 2
    while True:
        if q not in D:
            yield q
            D[q*q] = [q]
        else:
            for p in D[q]:
                D.setdefault(p+q, []).append(p)
            del D[q]
        q+=1
for i in gen_primes():
    if i > limit:
        break
    primes.append(i)
primes[desired-1]
```

## Problem 9

### Special Pythagorean Triplet

A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

$a^2 + b^2 = c^2$

For example, $3^2 + 4^2 = 9 + 16 = 25 = 5^2$

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.

```python
# This is a way to solve the problem, but not the most classy or efficient way
# The best way is by parameterization of the triplets
# a = m^2 - n^2, b = 2*m*n, c = m^2 + n^2
def GenTriplet():
    
    def Pythagorean(a, b):
        c = np.sqrt(a**2 + b**2)
        return(a, b, c)
    
    limit = 1000
    triplets = {}
    
    for i in range(1, limit+1):
        b = i
        for j in range(1, b):
            a = j
            c = Pythagorean(a, b)[2]
            if c.is_integer():
                A = Pythagorean(a, b)[0]
                B = Pythagorean(a, b)[1]
                C = int(c)
                triplets[C**2] = [A, B, C]
                if a+b+c==1000:
                    answer = [A, B, C, A*B*C]
   

    yield answer
```

## Problem 10

### Summation of Primes

The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
# This code gives the right answer, could be optimized I guess
limit = 2000000
primes = []
def GenPrimes():
    numbers = {}
    q = 2
    while True:
        if q not in numbers:
            yield q
            numbers[q*q] = [q]
        else:
            for p in numbers[q]:
                numbers.setdefault(q+p, []).append(p)
            del numbers[q]
        q+=1
    
for v in GenPrimes():
    if v > limit:
        break
    primes.append(v)
print(sum(primes))
```

## Problem 12

### Highly Divisible Triangular Number

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1  
 3: 1,3  
 6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?



```python
# Here is the best solution I could find.  I had to look online for advice because mine
# was taking so long.
%%time
from functools import reduce

def GenTriangle():
    '''
    Generates triangle numbers into infinity
    '''
    i=1
    n=0
    while True:
        n = n + i
        yield n
        i+=1
        
def factors(n):
    '''
    This will return all of the factors, very quickly, of a number n.

    Why square root as the upper limit?

    sqrt(x) * sqrt(x) = x. So if the two factors are the same, they're both the square root. 
    If you make one factor bigger, you have to make the other factor smaller. 
    This means that one of the two will always be less than or equal to sqrt(x), 
    so you only have to search up to that point to find one of the two matching factors. 
    You can then use x / fac1 to get fac2.

    The reduce(list.__add__, ...) is taking the little lists of [fac1, fac2], 
    and joining them together in one long list.

    The [i, n/i] for i in range(1, int(sqrt(n)) + 1) if n % i == 0 returns a pair of factors 
    if the remainder when you divide n by the smaller one is zero (it doesn't need to check 
    the larger one too; it just gets that by dividing n by the smaller one.)

    The set(...) on the outside is getting rid of duplicates, which only happens for 
    perfect squares. For n = 4, this will return 2 twice, so set gets rid of one of them.
    '''

    return set(reduce(list.__add__, 
                     ([i, n//i] for i in range(1, int(n**0.5) + 1) if n%i ==0))) 
# same as set(reduce(lambda x, y: x+y, ([i, n//i] for i in range(1, int(n**0.5) + 1) if n%i == 0)))
help(list.__add__)
help(reduce)

###################
dic = {}
divisors = 500
solution = None
for v in GenTriangle():
    dic[v] = factors(v)
    if len(factors(v)) > divisors:
        solution = v
        break
print(solution)
```

 ### Longest Collatz Sequence
 
 The following iterative sequence is defined for the set of positive integers:

$n â†’ n/2$ (n is even)

$n â†’ 3n + 1$ (n is odd)

$$13 â†’ 40 â†’ 20 â†’ 10 â†’ 5 â†’ 16 â†’ 8 â†’ 4 â†’ 2 â†’ 1$$
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
%%time
# Solution is 837799, takes about 55 seconds
limit = 1000000
longestchain = []
number = None
for seed in range(limit+1):
    n = seed
    l = [n]
    while n > 1:
        if n % 2 == 0:
            n = int(n/2)
            l.append(n)
        else:
            n = 3*n + 1
            l.append(n)
    if len(l) > len(longestchain):
        longestchain = l
        number = seed
print(number, len(longestchain))
```

## Problem 15

### Lattice Paths

Starting in the top left corner of a 2Ã—2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.


How many such routes are there through a 20Ã—20 grid?

```python
import math
grid = [20, 20] # columns, rows of grid
def permutations(n):
    """
    Takes a list of number of columns and rows ([m, n]) and returns 
    the number of possible paths through the lattice if the only possible movements
    are right and down
    """
    return int(math.factorial(sum(n))/(math.factorial(grid[0])*math.factorial(grid[1])))
print(f"number of possible paths: {permutations(grid)}")
```

```python
# Best answer!!!
# method of using a matrix and going from the bottom up
def LargestSumPath(string, OGmatrix=False):
    """
    Finds the path through a pyramid of numbers that gives the maximum sum.
    Turns the pyramid into a matrix by adding zeros.  Goes from the bottom of the pyramid
    to the top, testing for each number the number of the next line that gives
    the greatest sum.  The problem says the path needs to go from one number to 
    either of the adjacent numbers on the next line down.
    Args:
        string = str, preferably a docstring that is actually like a pyramid. Example:
            string = '''3
                        7 4
                        2 4 6
                        8 5 9 3'''
        OGmatrix = Boolean, if true it will return a matrix
    Returns:
        matrix[0][0] = int, the sum of the path
        OGmatrix (optional) = a matrix
    
    """
    # creating the matrix
    matrix = []
    rows = string.split('\n')
    for v in rows:
        line = []
        for n in v.split(' '):
            line.append(int(n))
        while len(line) < len(rows):
            line.append(0)
        matrix.append(line)
    # matrix is done
    # add from bottom up
    OGmatrix = matrix[:]
    counter = 0
    new = []
    for line in range(len(matrix)-1, 0, -1):
        temp = []

        for n in range(len(matrix[line])-1):
            if matrix[line-1][n] + matrix[line][n] > matrix[line-1][n] + matrix[line][n+1]:
                temp.append(matrix[line-1][n] + matrix[line][n])
                #print(matrix[line-1][n], matrix[line][n], matrix[line-1][n] + matrix[line][n])
            else:
                temp.append(matrix[line-1][n] + matrix[line][n+1])
                #print(matrix[line-1][n-1], matrix[line][n], matrix[line-1][n-1] + matrix[line][n])
        #while len(temp) < 4:
            #temp.append(0)
        new.append(temp)
        del(matrix[line])
        matrix[line-1] = temp
        
    counter+=1
    
    if OGmatrix==True:
        return matrix[0][0], OGmatrix
    else:
        return matrix[0][0]
```

### Amicable Numbers

Let d(n) be defined as the sum of proper divisors of n (numbers less than n which divide evenly into n).
If d(a) = b and d(b) = a, where a â‰  b, then a and b are an amicable pair and each of a and b are called amicable numbers.

For example, the proper divisors of 220 are 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 and 110; therefore d(220) = 284. The proper divisors of 284 are 1, 2, 4, 71 and 142; so d(284) = 220.

Evaluate the sum of all the amicable numbers under 10000.

```python
def isAmicable(a, b):
    
    if a == b:
        return False
    
    divisorsa = []
    divisorsb = []
    for i in range(1, a):
        if a%i == 0:
            divisorsa.append(i)
        if b%i == 0:
            divisorsb.append(i)
    
    if sum(divisorsa) == b and sum(divisorsb) == a:
        return True
    
    else:
        return False
```
